# 1. File Upload to Cloud.

Rather than manually uploading directly to the cloud I chose to do it via a script, to somewhat automate the process. Although I had to manually create the bucket.

```python
import os
import sys
from dotenv import load_dotenv
import warnings
```

```python
load_dotenv()
warnings.filterwarnings('ignore')
```

```python
current_dir = os.getcwd()
Root_dir = os.path.abspath(os.path.join(current_dir, '..'))

if Root_dir not in sys.path:
    sys.path.append(Root_dir)
```

```python
from src.File_Uploading_to_Cloud.Upload import get_gcs_client
from src.File_Uploading_to_Cloud.Upload import upload_csv
from src.File_Uploading_to_Cloud.Upload import ensure_bucket_exists
```

```python
BUCKET_NAME = os.getenv("BUCKET_NAME")
CREDENTIALS = os.path.join(Root_dir, os.getenv("CREDENTIALS_PATH"))
```

```python
client = get_gcs_client(credentials_path=CREDENTIALS)
ensure_bucket_exists(client, BUCKET_NAME)  
```

```python
# Uploading a csv file
data_path = "./data/World Crude Oil Reserves from 1995 to 2021.csv"
upload_csv(client, BUCKET_NAME, data_path)
```

```python
# Uploading a json file
data_path = "./data/worlds_busiest_container_ports.json"
upload_csv(client, BUCKET_NAME, data_path)
```

# 2. Files listing, reading, loading and download from Cloud. 

Authenticating, and retrieving files from the cloud.

```python
from src.File_Downloading_From_Cloud.Download import list_files, load_file, download_file, preview_file_as_json
```

```python
CREDENTIALS = os.path.join(Root_dir, os.getenv("CREDENTIALS_PATH"))
BUCKET_NAME = os.getenv("BUCKET_NAME")
```

```python
# Listing the files in the bucket
client = get_gcs_client(CREDENTIALS)
files = list_files(client=client, bucket_name=BUCKET_NAME)               
for file in files:
    print(file)
```

```python
# Previewing a file (Displays as a JSON)
result = preview_file_as_json(client, BUCKET_NAME,"world crude oil reserves from 1995 to 2021.csv")
print(result)
```

```python
# Loading any file for perhaps immediate manipulation without downloading from the cloud.
data = load_file(client=client, bucket_name=BUCKET_NAME, file_name = "world crude oil reserves from 1995 to 2021.csv")
data.head()
```

```python
# Downloading a cvs file
download_file(client=client, bucket_name= BUCKET_NAME, file_name = "world crude oil reserves from 1995 to 2021.csv")
```

```python
# Downloading a Json file
download_file(client=client, bucket_name=BUCKET_NAME, file_name = "worlds_busiest_container_ports.json")
```
